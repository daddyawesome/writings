```
#Nice Hex Spiral
import turtle
colors=['red','purple', 'blue','green', 'yellow', 'orange']

ts = turtle.getscreen()
ts.bgcolor('black')
for x in range(360):
  turtle.pencolor(colors[x%6])
  turtle.width(x/100+1)
  turtle.forward(x)
  turtle.left(59)
```

```
#Circle Spiral
import	turtle as t
#t	=	turtle.Pen() 
for	x	in	range(100):
  t.circle(x)
  t.left(91)
```

```
#Square 
import	turtle as t
t.pencolor("red")
for	x	in	range(100):
  t.forward(x)
  t.left(91)
```
